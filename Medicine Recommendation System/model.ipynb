```python
# Train test split 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
# Here we have to pre-process the prognosis
```

```python
X = dataset.drop('prognosis',axis=1)
# X is the input column so we drop the prognosis
y = dataset['prognosis']
# y is the output column 
```

```python
X
```

```python
# We use label encoder to convert the string value into numerical value 
# because machine cann't understand the string thing 
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
Y
```

```python
unique_vals = np.unique(Y)
print(unique_vals)
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,Y,test_size=0.3, random_state=20)
# X_train - Training features (70% of your input data)
# X_test - Testing features (30% of your input data)
# y_train - Training labels (70% of your output data)
# y_test - Testing labels (30% of your output data) 
# random_state=20 - Sets a seed for reproducible results
# (you'll get the same split every time)
```

```python
X_train.shape , X_test.shape ,y_train.shape , y_test.shape
```

```python
# Training Top Models 

from sklearn.datasets import make_classification
# Used to generate a synthetic classification dataset.

from sklearn.svm import SVC
# Support Vector Classifier is  powerful classifier for binary/multi-class 
# classification problems,especially with clear margin of separation.

from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score,confusion_matrix
# accuracy_score - Calculates the ratio of correctly predicted instances to the total instances
# confusion_matrix - Creates a matrix showing the counts of true positives, false positives, true negatives, and false negatives.
import numpy as np
```

```python
# Create a dictionay to store the models 
models = {
    "SVC":SVC(kernel='linear'),
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNeighbors":KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB":MultinomialNB()
}

# Here we use kernel linear as most of the data is linearly seperable.
# n_estimators=100 - Use 100 decision trees in the random forest
# andom_state=42 - Ensure consistent results across runs by fixing randomness
```

```python
for model_name,model in models.items():
    # Train Model
    model.fit(X_train,y_train)
    
    # Evaluate the model
    predictions = model.predict(X_test)
    
    # Calculate Accuracy
    accuracy = accuracy_score(y_test,predictions)
    
    # Calculate Confussion Matrix
    cm = confusion_matrix(y_test,predictions)
    # What is Confussion Matrix
    # It shows how well your classification model is doing by 
    # comparing the actual (true) labels with the predicted labels.
    
    # Print the results
    print(f"{model_name} Accuracy: {accuracy}")
    print(f"{model_name} Confussion Matrix:")
    print(np.array2string(cm,separator=', ')) 
    # cm will be an array so we have convert it into string using np.array2string
```

```python
# Single Predictions 

svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```
